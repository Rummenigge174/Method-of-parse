```python
from bs4 import BeautifulSoup as bs
import requests
from pprint import pprint
```

```python
s = input('Введите название вакансии:' )
l = s.split()
print(l)
text = '+'.join(l)
print(text)
```

```python
#text='Data+scientist'
pages_fact = 0
pages = 0
vacancy_get = []
while pages == pages_fact:
    header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36'}
    main_link = 'https://hh.ru'
    response = requests.get(main_link + f'/search/vacancy?L_is_autosearch=false&area=113&clusters=true&enable_snippets=true&text={text}&page={pages}',headers=header).text
    soup = bs(response,'lxml')
    vacancy_block = soup.find('div',{'class':'vacancy-serp'})
    vacancy_list = vacancy_block.find_all('div',{'data-qa':'vacancy-serp__vacancy'})
    pages_butt = soup.find('div',{'class':'bloko-column bloko-column_l-13 bloko-column_m-9 bloko-column_s-8 bloko-column_xs-4'})
    flag = pages_butt.find('a',{'class':'bloko-button HH-Pager-Controls-Next HH-Pager-Control'})
    
    for vacancy in vacancy_list:
        vacancy_data = {}
        vacancy_name = vacancy.find('a',{'class':'bloko-link HH-LinkModifier'}).getText()
        vacancy_link = main_link + vacancy.find('a',{'class':'bloko-link HH-LinkModifier'})['href']
        vacancy_income = vacancy.find('div',{'class':'vacancy-serp-item__sidebar'}).getText()
        if vacancy_income:
            vacancy_income = vacancy_income
        else:
            vacancy_income = None
        vacancy_data['Название вакансии'] = vacancy_name
        vacancy_data['Зарплата'] = vacancy_income
        vacancy_data['Ссылка на вакансию'] = vacancy_link
        vacancy_data['Сайт'] = 'hh.ru'
        vacancy_get.append(vacancy_data)
    if flag:
        pages_fact += 1
    pages +=1
    

    
    
```

```python
print(len(vacancy_get))
```

Сплит зарплаты

```python
for i in range(0, len(df['Зарплата'])):
    s = df['Зарплата'][i]
    if (s is not None) and ('-' in s):
        s1 = s.split('-')
        val = s1[-1].split()
        min_zp = s1[0].replace('\xa0', '')
        min_zp = int(min_zp)
        max_zp = []
        for i in range(0, (len(val))):
            max_zp.append(val[i])
        del max_zp[-1]
        max_zp = ''.join(max_zp)
        max_zp = int(max_zp)
        currency = val[-1]
        df['Мин'][i] = min_zp
        df['Макс'][i] = max_zp
        df['Валюта'][i] = currency
    elif (s is not None) and ('-' not in s):
        s1 = s.split()
        min_zp = s1[1] + s1[2]
        min_zp = int(min_zp)
        max_zp = None
        currency = s1[-1]
        df['Мин'][i] = min_zp
        df['Макс'][i] = max_zp
        df['Валюта'][i] = currency
    else:
        min_zp = None
        max_zp = None
        currency = None
        df['Мин'][i] = min_zp
        df['Макс'][i] = max_zp
        df['Валюта'][i] = currency

```
