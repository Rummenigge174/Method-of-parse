```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
import time
import json
from pprint import pprint
import requests
from lxml import html
```

```python
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'}
driver = webdriver.Chrome('/Users/ars/GU_AI/GU_Methods_of_parse/lesson_5/chromedriver')
driver.get('https://mail.ru/')
mainlink = 'https://e.mail.ru'
i=0
k=0

letters=[]
my_letters = []
links = []
elem = driver.find_element_by_id('mailbox:login')
elem.send_keys('study.ai_172@mail.ru')
elem.send_keys(Keys.ENTER)
time.sleep(1)
elem = driver.find_element_by_id('mailbox:password')
elem.send_keys('NextPassword172')

elem.send_keys(Keys.ENTER)
time.sleep(10)

# –î–µ–ª–∞–µ–º —Å–∫—Ä–æ–ª–ª–∏–Ω–≥ –¥–æ —Å–∞–º–æ–≥–æ –¥–Ω–∞ —Å–æ–±–∏—Ä–∞—è –≤—Å–µ —Å—Å—ã–ª–∫–∏ –≤ —Å–ø–∏—Å–æ–∫ links
for i in range(15):
    time.sleep(1)
    set_letters = driver.find_element_by_xpath("//div[@class='dataset__items']")
    letters = set_letters.find_elements_by_class_name('llc')
    for link in letters:
        link = link.get_attribute('href')
        links_set = set(links)
        links_set.add(link)
        links = list(links_set)
    actions = ActionChains(driver)
    actions.move_to_element(letters[-1])
    actions.perform()
print(len(links))

# –ü–µ—Ä–µ–±–∏—Ä–∞–µ–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ links —Å –ø–µ—Ä–µ—Ö–æ–¥–æ–º –ø–æ –∫–∞–∂–¥–æ–º—É —ç–ª–µ–º–µ–Ω—Ç—É —Å–ø–∏—Å–∫–∞(—Å—Å—ã–ª–∫–µ) 
for letter in links:
    driver.get(letter)
    time.sleep(2)
    try:
        item_dict = {}
        letter_id = driver.find_element_by_class_name('thread__letter')
        from_letter = driver.find_element_by_class_name('letter-contact')
        date_letter = driver.find_element_by_class_name('letter__date')
        topic_letter = driver.find_element_by_class_name('thread__subject')
        body_letter = driver.find_element_by_class_name('letter__body')
        item_dict['_id'] = letter_id.get_attribute('data-id')                                              
        item_dict['From'] = from_letter.get_attribute('title')
        item_dict['Date'] = date_letter.text
        item_dict['Topic'] = topic_letter.text
        item_dict['Body'] = body_letter.text
        my_letters.append(item_dict)

    except NoSuchElementException:
        pass
        
    time.sleep(2)
    driver.back()
    time.sleep(2)
driver.quit()
pprint(my_letters)
```
