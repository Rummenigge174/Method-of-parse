```python
from pprint import pprint
from lxml import html
import requests
import re
import datetime
import json
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
chrome_options = Options()
chrome_options.add_argument('start-maximized') 
import time
header_h2=[]

header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'}
driver = webdriver.Chrome('/Users/ars/GU_AI/GU_Methods_of_parse/lesson_5/chromedriver', options=chrome_options)
driver.get('https://www.mvideo.ru/')
time.sleep(5)
response = requests.get('https://www.mvideo.ru/', headers=header)
dom = html.fromstring(response.text)
news = []
k=0
j=0
z=0
# Ищем элементы с классом h2, затем перебираем список и ищем совпадение со строй 'Хиты продаж'
time.sleep(2)
elem = driver.find_elements_by_class_name('h2')
items = []
for i in elem:
    elem_text = i.text
    if 'Хиты продаж' in elem_text:
        elem_hit =i

# Найдя нужный блок, поднимаемся от h2 до нужного div и собираем данные.
while j < 4:
    try:
        goods = elem_hit.find_elements_by_xpath("../../../../div[@class='section']//li")
        print('lenght = ',len(goods))
        for good in goods:
            item = {}
            name = good.find_elements_by_class_name('e-h4')
            link = good.find_elements_by_class_name('sel-product-tile-title')
            price = good.find_elements_by_class_name('c-pdp-price__current')
            if (name[0].text != ''):
                item['Price'] = price[0].text.replace('¤', ' руб.')
                item['Name'] = name[0].text
                item['Link'] = link[0].get_attribute('href')
                items.append(item)
        time.sleep(2)
        button = elem_hit.find_element_by_xpath("../../../../div[@class='section']//a[@class='next-btn sel-hits-button-next']")
        button.click()
        j += 1
        z += 1
        time.sleep(3)
    except:
        print('Button_count_pressed ',j)
        break
print(j)
print ('count in for',z)

pprint(items) 

```
